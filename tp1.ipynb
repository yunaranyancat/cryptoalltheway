Hum... Ce message contient un mélange de caractères exotiques (cyrilliques) et de ponctuation. Commençons par isoler uniquement les caractères sur lesquels le chiffrement a été appliqué.

```python
s=''.join([x for x in c if x.isalpha()])
print(s)
```

Comptons le nombre d'occurence de chaque caractère dans ce texte :

```python
freq={}
for x in s:
    if x in freq:
        freq[x]=freq[x]+1
    else:
        freq[x]=1
print(freq)
```

```python
def ic(freq):
    """TODO: écrire ici la fonction qui calcule l'IC à partir du dictionnaire de fréquences freq
             pour l'instant cette fonction retourne le nombre total de caractères."""

     
    n=0
    ic=0
    for x in freq:
        nx=freq[x]  # nombre d'occurences du caractère x
        ic=ic+nx*(nx-1)
        n=n+nx
    return ic/(n*(n-1))

print(ic(freq))
```

Voici quelques fonctions qui devraient vous aider :

```python
def tolet(x):
    "transforme l'entier [x] en la [x]ième lettre majuscule (modulo 26)"
    return chr(65+x%26)
    
def toint(c):
    "tranforme la lettre majuscule [c] en sa position dans l'alphabet (de 0 à 25)"
    return ord(c)-65

def shift(c,d):
    """décale la lettre [c] de k positions dans l'alphabet (modulo 26)
     où [k] est la position de la lettre [d] dans l'alphabet"""
    return tolet(toint(c)+toint(d))
```

```python
from unicodedata import normalize

def nettoie(m):
    "nettoie la chaîne [m] en ne gardant que les lettres débarassées de leurs accents et passées en majuscules"
    return ''.join([c for c in normalize('NFKD',m).upper().strip() if ord(c)>64 and ord(c)<91])
```

```python
sx=nettoie("poule")
print(sx)
sy=nettoie("lapin")
print(sy)
print("-----")
# décale le mot [sx] en utilisant les lettres successives de [sy] comme valeurs de décalage
print(''.join([shift(sx[i],sy[i]) for i in range(len(sx))]))
```

```python
def extraireSousTextes(m,k):
    lesSousTextes = [ "" for i in range(k)]
    for i in range(len(m)):
        lesSousTextes[i%k] += m[i]
    return lesSousTextes

print(extraireSousTextes(m,6))
```

<div class="alert alert-info">**Q4** Définir une fonction ``calculICMoyen`` permettant de calculer l'IC moyen d'une liste de n textes.</div>

```python
def freq(m):
    f={}
    for c in m:
        f[c] = f.get(c,0) + 1
    return f

def icm(m):
    return ic(freq(m))

def calculICMoyen(textes):
    ics=0
    for t in textes:
        ics+=icm(t)   
    return ics/len(textes)

for i in range(1,10):
    print(i,calculICMoyen(extraireSousTextes(m,i)))
```

<div class="alert alert-info">**Q5** Écrire la fonction ``trouveLeE`` qui retournera un entier correspondant au décalage vers la droite modulo 26. Par exemple vous trouvez que le ``E`` semble être la lettre ``C``. Le décalage est donc égal à 24 _i.e._ ``toint('E')+24 = (4+24)%26 = 2 = toint('C')``.  Définir la fonction ``calculeDecalage`` qui permet en fonction d'un texte donné de calculer le décalage semblant avoir été fait dans cet alphabet _i.e._ analyse des fréquences.</div>



```python
def trouveLeE(texte):
    noc=0
    lee=None
    f=freq(texte)
    for c in f:
        if f[c]>noc:
            noc=f[c]
            lee=c
    return lee

def calculerDecalage(texte):
    return (toint(trouveLeE(texte))-toint('E'))%26

print(calculerDecalage('CCCCCCCCCCCCCCCCCCCCCCCCCC'))
```

<div class="alert alert-info">**Q9** Définir la fonction ``chiffrementVigenere`` qui à partir d'un texte et d'une clé va chiffrer le texte en fonction de cette clé.</div>

```python
def chiffrementVigenere(texteClair,cle):
    chiffre = ""
    for i in range(len(texteClair)):
        chiffre += shift(texteClair[i],cle[i%len(cle)])
    return chiffre

print(chiffrementVigenere("BONJOUR","JEANNE"))
print(chiffrementVigenere("KSNWBYA","RWANNW"))
```

```python
def decrypt(ciphertext, key):
    key_length = len(key)
    key_as_int = [ord(i) for i in key]
    ciphertext_int = [ord(i) for i in ciphertext]
    plaintext = ''
    for i in range(len(ciphertext_int)):
        value = (ciphertext_int[i] - key_as_int[i % key_length]) % 26
        plaintext += chr(value + 65)
    return plaintext

```
